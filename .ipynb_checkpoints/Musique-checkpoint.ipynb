## Filtrage audio sous python

Au lieu de filtrer un signal avec une méthode analogique (comme vue en cours : RC, RLC, filtres en cascade), il est possible d'appliquer un filtre numérique quelconque à un signal enregistré. C'est une méthode bien moins onéreuse et plus polyvalente (changer une valeur dans un programme est bien plus aisée que de changer un composant électronique dans un circuit)

Dans  un premier temps, un test son est effectué : une note (la$_3$) à est jouée pendant 3 secondes. Vous pouvez changer les paramètres sans crainte d'impacter ce qui suit. Si vous n'entendez pas le son, peut-être faut-il relancer la cellule ou le noyau Python.

Dans un deuxième temps, un fichier audio est importé et joué. J'espère que vous garderez bien ce son en tête... Le but est de séparer les sons graves des sons aigus présents dans le morceau en lui appliquant des filtres passe bas et passe haut. 

## Test son

```python
import numpy as np
from scipy.fft import rfft, rfftfreq, fft, ifft
from scipy.io.wavfile import read
import matplotlib.pyplot as plt
import IPython

%matplotlib widget

frequence = 440  # Fréquence de la note jouée
ech = 44100  # nombre d'échantillons par secondes (joue sur la qualité des sons aigus)
duree = 1  # Durée de la note : 3 secondes

# Générer un tableau de valeur avec duree*ech pas, de 0 à duree secondes
t = np.linspace(0, duree, duree * ech, False)

# Générer un tableau représentant un signal sinusoïdal pendant duree secondes
note = np.sin(frequence * t * 2 * np.pi)

# Fixer le maximum du signal sur un interval de 16 bit
#note = note * (2**15 - 1) / np.max(np.abs(note))
# Convertir chaque valeur du tableau en des données à 16 bit
#note = note.astype(np.int16)

# Tracé du signal sinusoïdal 
plt.figure(figsize=(10,5))
plt.plot(t,note)
plt.xlim([0,0.1])
plt.show()

# Lecture audio du signal sinusoïdal
IPython.display.Audio(note,rate=ech, autoplay=True)
```

## Filtrer un morceau de musique

```python
filename = 'noel.wav'
contenu = read(filename)
ech = 22050
morceau = np.array(contenu[1],dtype=float)
morceau = morceau[:,1] #steréo -> mono
L = np.size(morceau)
t = np.linspace(0,L/ech,L, False)
plt.figure()
plt.plot(t,morceau)
plt.xlabel('Temps (s)')
plt.ylabel('Amplitude (bit)')
plt.title('Chronogramme du morceau de musique')
plt.show()
IPython.display.Audio(filename,rate=ech, autoplay=True)
```

Le morceau de musique peut être filtré pour séparer les hautes et basses fréquences, pour séparer le son provenant de chaque instrument par exemple.

### Création d'un filtre passe bas

```python
taille = np.size(fft(morceau))
FiltrePB = []
freq = rfftfreq(L, t[1] )
fc = 430
for f in freq:
    if f < fc :
        FiltrePB.append(1)
    else:
        FiltrePB.append(1e-20)
FiltrePB = FiltrePB[1:]+FiltrePB[-1:0:-1]
FiltrePB = np.array(FiltrePB)

plt.figure(figsize=(10,6))
plt.plot(freq,20*np.log10(FiltrePB[0:taille//2+1]))
plt.xscale("log")
plt.ylim([-80,10])
plt.xlabel('Fréquence (Hz)')
plt.ylabel('Gain du filtre (dB)')
plt.title('Filtre passe bas de fréquence de coupure = {:.0f} Hz'.format(fc))
plt.show()
```

### Application du filtre passe bas sur le morceau de musique

```python
plt.figure(figsize=(10,6))
plt.plot(freq,np.abs(fft(morceau))[0:taille//2+1]*2/taille,label='signal non filtré')
plt.plot(freq,np.abs(FiltrePB*fft(morceau))[0:taille//2+1]*2/taille,label='signal filtré')
plt.yscale("log")
plt.xscale("log")
plt.ylim([1e-2,1e4])
plt.xlabel('Fréquence (Hz)')
plt.ylabel('Amplitude du signal (bit)')
plt.title('Transformée de Fourier du morceau de musique avant et après filtrage')
plt.legend()
plt.show()

plt.figure(figsize=(10,6))
plt.plot(t,morceau,label='signal non filtré')
plt.plot(t,np.real(ifft(FiltrePB*fft(morceau))),label='signal filtré')
plt.xlabel('Temps (s)')
plt.ylabel('Amplitude du signal (bit)')
plt.title('Chronogramme du morceau de musique avant et après filtrage')
plt.legend()
plt.show()

sons_graves = np.real(ifft(FiltrePB*fft(morceau)))
sons_graves = sons_graves * (2**15 - 1) / np.max(np.abs(sons_graves))

IPython.display.Audio(sons_graves,rate=ech, autoplay=True)
```

La transformée de Fourier du morceau de musique fait apparaître les différentes fréquences des notes jouées dans le morceau. Les fréquences des notes jouées semblent être contenues entre 80 Hz et 2 kHz environ (amplitude dépassant les 10$^2$ bits).

Le signal filtré correspond au signal d'entrée de 0 à 430 Hz (gain = 0 dans la bande passante du filtre). Les fréquences plus élevées que la fréquence de coupure sont fortement atténuées. Vous pouvez modifier cette atténuation en modifiant le nombre `1e-20` contenu dans la commande `FiltrePB.append(1e-20)` par exemple en `1e-1` et écouter le résultat.  

### Création d'un filtre passe haut

```python
FiltrePH = []
fc = 2500
for f in freq:
    if f > fc :
        FiltrePH.append(1)
    else:
        FiltrePH.append(1e-20)
FiltrePH = FiltrePH[1:]+FiltrePH[-1:0:-1]
FiltrePH = np.array(FiltrePH)

plt.figure(figsize=(10,6))
plt.plot(freq,20*np.log10(FiltrePH[0:taille//2+1]))
plt.xscale("log")
plt.ylim([-80,10])
plt.xlabel('Fréquence (Hz)')
plt.ylabel('Gain du filtre (dB)')
plt.title('Filtre passe haut de fréquence de coupure = {:.0f} Hz'.format(fc))
plt.show()
```

### Application du filtre passe haut sur le morceau de musique

```python
plt.figure(figsize=(10,6))
plt.plot(freq,np.abs(fft(morceau))[0:taille//2+1]*2/taille,label='signal non filtré')
plt.plot(freq,np.abs(FiltrePH*fft(morceau))[0:taille//2+1]*2/taille,label='signal filtré')
plt.yscale("log")
plt.xscale("log")
plt.ylim([1e-2,1e4])
plt.show()
plt.figure(figsize=(10,6))
plt.plot(t,morceau,label='signal non filtré')
plt.plot(t,np.real(ifft(FiltrePH*fft(morceau))),label='signal filtré')
plt.legend()
plt.show()

sons_aigus = np.real(ifft(FiltrePH*fft(morceau)))
sons_aigus = sons_aigus * (2**15 - 1) / np.max(np.abs(sons_aigus))

IPython.display.Audio(sons_aigus,rate=ech, autoplay=True)
```
